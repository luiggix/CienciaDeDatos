```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importo las librerías necesarias

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df = pd.read_csv(url)
```

Cargo la base de datos actualizada directamente desde el GitHub

```python
df.head()
```

Muestro un resumen de los datos

```python
paises = ['Mexico','Singapore', 'Argentina', 'Chile', 'Korea, South', 'Japan', 'Ecuador', 'Germany', 'US', 'Russia']
```

Selecciono los países de interés mediante una lista con su nombre en la base de datos

```python
def getInfectedByCountry(country, startDate = '1/22/20', dropZeros = True):
    lista_pais = list(df[df['Country/Region'] == country].loc[:,startDate:].sum())
    if dropZeros:
        return np.trim_zeros(lista_pais)
    else:
        return lista_pais
```

Esta función nos da la lista de casos ya sumando todos los renglones del país, con la opción por defecto de empezar en los días con casos mayores a cero.

```python
fig, ax = plt.subplots()
plt.rcParams.update({'font.size': 14})
for p in paises:
    if p == 'Mexico':
        plt.plot(getInfectedByCountry(p),label=p, alpha = 0.8, linewidth=5, marker='o', markersize=5)
    else:
        plt.plot(getInfectedByCountry(p),label=p, alpha = 0.5, linewidth=3, marker='o', markersize=3)
plt.ylim(0,10000)
fig.set_size_inches(15,8)
plt.legend()
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.set_xlabel('Días desde el primer caso')
ax.set_ylabel('Infectados')
ax.set_title('Casos confirmados de Covid-19')
```

Simplemente se itera la lista de paises de interés, ploteando los datos obtenidos de la función. Para el país de interés (México) se realiza una línea más gruesa.

```python
def calculate_e1(time_series):
    salida = []
    for e in range(len(time_series)-1):
        salida.append(time_series[e+1]/time_series[e])
    return np.average(salida)
```

Esta función calcula el promedio de los E+1 calculados en la serie de tiempo dada.

```python
def getExponentialModelSeries(startValue,e1,days):
    predicted_series = [startValue]
    for i in range(1,days):
        predicted_series.append(predicted_series[-1]*e1)
    return predicted_series
```

Esta función nos permite darle el E+1 calculado, y el valor de inicio. En base a estos y el número de días a calcular, nos devuelve la predicción del modelo exponencial.

```python
def getMSE(time_series, proposal):
    predicted_series = getExponentialModelSeries(time_series[0],proposal,len(time_series))
    return np.sum((np.array(predicted_series)-np.array(time_series))**2)
```

A esta función se le pasa la serie de tiempo real y un E+1 propuesta. Entonces se genera una muestra del mismo tamaño a la muestra real, y se devuelve el error cuadrático medio de la propuesta en comparación a los casos reales.

```python
def adjust_e1_model(country,days=-1):
    time_series = getInfectedByCountry(country)[:days]
    bestTime = 1
    bestE1 = calculate_e1(time_series)
    bestMSE = getMSE(time_series,bestE1)
    for i in range(len(time_series)):
        proposal = calculate_e1(time_series[i:])
        proposalMSE = getMSE(time_series,proposal)
        if proposalMSE < bestMSE:
            bestTime = i+1
            bestE1 = proposal
            bestMSE = proposalMSE
    return (bestTime,time_series[bestTime],bestE1,bestMSE)
```

Esta función nos permite ajustar el modelo exponencial.
Se le pasa el país de interés y el número de días desde el primer caso para el que se quiere ajustar el modelo.

Simplemente se itera en torno a los datos reales para ver si se inicia del día $i$, cuál es el error cuadrático medio de la predicción E+1. Se devuelve el día del que se inició la mejor predicción, el número de casos que había en ese día, la predicción E+1 y el error cuadrático medio de la predicción.

```python
dias_a_predecir = 30
dias_a_mostrar = 30
(tiempoInicio, casos, e1, error) = adjust_e1_model('Mexico',days=dias_a_predecir)
```

Damos los días a mostrar en el gráfico, el número de días para el que queremos ajustar nuestro modelo desde que se encontró el primer caso, y ajustamos el modelo en cuestión.

```python
fig, ax = plt.subplots()
casosReales = getInfectedByCountry('Mexico')[:dias_a_predecir]
plt.plot(casosReales, label='Casos Reales en México', alpha = 0.8, linewidth=0, marker='o', markersize=5)
plt.plot(getExponentialModelSeries(casosReales[0],e1,dias_a_mostrar), label='Modelo para México')
plt.xlim(0,dias_a_mostrar)
fig.set_size_inches(15,8)
plt.legend()
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.set_xlabel('Días desde el primer caso')
ax.set_ylabel('Infectados')
ax.set_title('Covid-19: Modelo para predecir los primeros '+str(dias_a_predecir)+' días ajustando desde los '+ str(casos) +' casos (E+1=' +str(np.round(e1,2)) +')')
```

Graficamos tanto el modelo como los casos reales. En el título se muestra el rango de días para el que se ajustó el modelo, y el número de casos desde el cuál se encontró el modelo óptimo. También se muestra el crecimiento E+1.
