```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
# Leemos los datos descargados de https://github.com/CSSEGISandData/COVID-19
#covid_19 = pd.read_csv('COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
covid_19 = pd.read_csv(url)
covid_19.head()
```

```python
# Seleccionamos 10 paises de nuestra preferencia.
#covid_19_10 = covid_19[covid_19.isin(["Mexico","New Zealand", "Canada", "Portugal", "Norway", "Chile", "Belgium", "Argentina", "Denmark", "Netherlands"]).loc[:,"Country/Region"]]
covid_19_10 = covid_19[covid_19.isin(["Mexico","New Zealand", "Canada", "Portugal", "Norway", "Italy", "Belgium", "Spain", "China", "Netherlands"]).loc[:,"Country/Region"]]

# Quitamos las columnas innecesarias
covid_19_10 = covid_19_10.drop(columns = 'Province/State Lat Long'.split())
# Hacemos nuevo dataframe sin los estados, sólo los paises
regiones = covid_19_10['Country/Region'].unique()
covid = pd.DataFrame(columns = covid_19_10.columns[1:], index = regiones)
# Llenamos el dataframe con la suma para todos los paises que se eligieron
for region in regiones:
    covid.loc[region, :] = covid_19_10[covid_19_10['Country/Region'].isin([region])].sum()
covid = covid.loc[:,covid.sum() > 0]
covid
```

```python
# Definimos el eje x para la graficación
x = range(covid.shape[1])
# Lista que contiene los diferentes estilos de línea. 
ls = ['dashdot', (0, (5, 5)), (0, (1, 1)), (0, (5, 10)), (0, (1, 1)), 
      (0, (5, 1)), (0, (3, 1, 1, 1, 1, 1)), (0, (3, 5, 1, 5)), (0, (3, 1, 1, 1)), (0, (3, 5, 1, 5, 1, 5))]
```

# Modelo de regresión exponencial.

```python
# Aquí se definen unas variables para los colores de las gráficas. 
j = 0
color = ['r', 'k']
# Se inicializa la gráfica
fig, ax = plt.subplots(figsize = (20, 10))
# Ciclo for para graficar cada país objetivo
for i, idx in enumerate(covid.index):
    # Aquí definimos en qué país nos queremos fijar y hacer su modelo exponencial
    if idx == 'Mexico':# or idx == 'New Zealand':
        # Graficamos individualmente al país objetivo
        # Notemos que se quitan todos los ceros con la función "np.trim_zeros" de numpy
        ax.plot(np.trim_zeros(covid.loc[idx, :].values), lw = 2, ls = ls[i], color = color[j], 
                label = idx, marker = "X", ms = 10)
        
        #---------------------------------------------------------------------------------------
        # Regresión exponencial:
        y = np.trim_zeros(covid.loc[idx, :].values[2:])
        X = np.arange(len(y))
        X = X.reshape(-1,1)
        y = np.log(list(y))
        reg = LinearRegression()
        reg.fit(X, y)
        β_1 = np.exp(reg.coef_[0])
        β_0 = np.exp(reg.intercept_)
        X = np.arange(55)
        ax.plot(β_0*((β_1)**X), lw = 3, label = f'Modelo exponencial: {idx}', color = color[j])
        #----------------------------------------------------------------------------------------
        
        j += 1
    else:
        # Graficamos los otros paises con un degradado para que no se lleven el protagonismo
        ax.plot(np.trim_zeros(covid.loc[idx, :].values), lw = 3, ls = ls[i], alpha = 0.6, label = idx)
        pass

# Ajustamos unos parámetros de las gráficas
ax.set_title("Crecimiento exponencial", fontsize = 40)
ax.legend(fontsize = 18)
ax.set_ylabel('Casos reportados', fontsize = 30)
ax.set_xlabel('Días transcurridos', fontsize = 30)
xt = np.linspace(0, 80, 21).astype(int)
yt = np.arange(0, 136675, 15000).astype(int)
ax.set_xticks(xt)
ax.set_xticklabels(xt, fontsize = 20)
ax.set_yticks(yt)
ax.set_yticklabels(yt, fontsize = 20)
ax.grid(True, alpha = 0.1)
ax.set_xlim((0,76))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.text(36, 135000, f'(E+1) = {β_1}', 
            size = 25, rotation = 0, ha = "center", va = "center",
            bbox=dict(boxstyle="round", ec=(1., 0.5, 0.5), fc=(1., 0.8, 0.8),))
plt.show()
```

# Modelo de regresión exponencial lineal

```python
# Aquí se definen unas variables para los colores de las gráficas. 
j = 0
color = ['r', 'k']
betas = []
# Se inicializa la figura en donde se graficará
fig, ax = plt.subplots(figsize = (20, 10))
ax.set_yscale("log", nonposy='clip')
# Ciclo for para graficar cada país objetivo
for i, idx in enumerate(covid.index):
    # Aquí definimos en qué país nos queremos fijar y hacer su modelo exponencial
    if idx == 'Mexico' or idx == 'New Zealand':
        # Graficamos individualmente al país objetivo
        # Notemos que se quitan todos los ceros con la función "np.trim_zeros" de numpy
        ax.plot(np.trim_zeros(covid.loc[idx, :].values), lw = 2, ls = ls[i], color = color[j], 
                label = idx, marker = "X", ms = 10)
        #----------------------------------------------------------------------------------------
        #Regresión:
        y = np.trim_zeros(covid.loc[idx, :].values[2:])
        X = np.arange(len(y))
        X = X.reshape(-1,1)
        y = np.log(list(y))
        reg = LinearRegression()
        reg.fit(X, y)
        β_1 = np.exp(reg.coef_[0])
        betas.append(β_1)
        β_0 = np.exp(reg.intercept_)
        X = np.arange(55)
        ax.plot(β_0*(β_1**X), color = color[j], lw = 3, label = f'Modelo exponencial: {idx}')
        #----------------------------------------------------------------------------------------
        
        j += 1
    else:
        # Graficamos los otros paises con un degradado para que no se lleven el protagonismo
        ax.plot(np.trim_zeros(covid.loc[idx, :].values), lw = 3, ls = ls[i], alpha = 0.6, label = idx)

# Ajustamos unos parámetros de las gráficas
# Ajustamos unos parámetros de las gráficas
ax.set_title("Crecimiento exponencial", fontsize = 40)
ax.legend(fontsize = 18)
ax.set_ylabel('Casos reportados', fontsize = 30)
ax.set_xlabel('Días transcurridos', fontsize = 30)
xt = np.linspace(0, 80, 21).astype(int)
ax.set_xticks(xt)
ax.set_xticklabels(xt, fontsize = 20)
ax.grid(True, alpha = 0.1)
ax.set_xlim((0,76))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.text(33, 35000, f'(E+1) = {betas[0]}', 
            size = 20, rotation = 0, ha = "center", va = "center",
            bbox=dict(boxstyle="round", ec=(1., 0.5, 0.5), fc=(1., 0.8, 0.8),))
ax.text(65, 35000, f'(E+1) = {betas[1]}', 
            size = 20, rotation = 0, ha = "center", va = "center", color = 'white',
            bbox=dict(boxstyle="round", ec=(1., 1., 1.), fc=(0., 0., 0.),))
plt.show()
```
